```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import ImageGrid

f_dir = 'C:\\Users\\ADMIN\\Desktop\\ENVI\\output\\'

f1_s = '\\SUMMER\\ATMO\\exp1_summer_AT_2018-12-21_'
f2_s = '.00.01.nc'

variables = ['Flowu',
             'Flowv',
             'Floww',
             'WindSpeed']

raw = nc.Dataset(f_dir + f1_s + "07" + f2_s)
data = [raw[var][0,:,:] for var in variables] 
```

```python
fig, axes = plt.subplots(1, len(variables), figsize = (30,5))
for i, dt in enumerate(data):
    cmap = plt.get_cmap('jet')
    var_plot = axes[i].pcolormesh(dt, cmap = cmap)
    axes[i].set_title(variables[i])
    plt.colorbar(var_plot)
    
fig.subplots_adjust(right = 0.8)
fig.suptitle('7AM' )


```
